## Data cleaning steps

```python
import re

# Rename columns to lowercase with underscores
def to_snake(name):
    s = re.sub(r"\s+", " ", str(name)).strip().lower()
    s = re.sub(r"[^0-9a-z]+", "_", s)
    return re.sub(r"_+", "_", s).strip("_")

df.columns = [to_snake(c) for c in df.columns]
df.head(1)
```

```python
# Trimming spaces and replace blank string with NaN
for col in df.select_dtypes(include="object"):
    df[col] = df[col].astype(str).str.strip().replace({"nan": np.nan, "": np.nan})
```

## Simplifying Key Columns

```python
# Simplify Race/Ethnicity
def simplify_race(val):
    if pd.isna(val): return np.nan
    s = str(val).lower()
    if ";" in s: return "multi_racial"
    if "white" in s or "caucasian" in s: return "white"
    if "hispanic" in s or "latino" in s: return "hispanic_latino"
    if "black" in s or "african" in s: return "black_african_american"
    if "asian" in s: return "asian"
    if "native" in s or "american indian" in s: return "native_american"
    return "other"

if "race_ethnicity" in df.columns:
    df["race_simple"] = df["race_ethnicity"].apply(simplify_race)
```

```python
# Simplify Gender
def simplify_gender(val):
    if pd.isna(val): return np.nan
    s = str(val).lower()
    if "man" in s or s == "male": return "man"
    if "woman" in s or s == "female": return "woman"
    if "non" in s or "nb" in s: return "nonbinary"
    return "other"

if "gender" in df.columns:
    df["gender_simple"] = df["gender"].apply(simplify_gender)
```

## Focus Columns and Save the Cleaned File

```python
# Selecting the key columns to keep
cols_to_keep = [
    "timestamp",
    "what_degree_program_are_you_currently_enrolled_in",
    "gender_simple",
    "race_simple",
    "age"
]
df_clean = df[cols_to_keep].copy()

# Droppings the rows with missing course info
df_clean = df_clean.dropna(subset=["what_degree_program_are_you_currently_enrolled_in"])

# Saving the cleaned csv file
df_clean.to_csv("majors_survey_cleaned.csv", index=False)
df_clean.head()
```
