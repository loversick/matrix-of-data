### Pulling my current playlists from Spotify and storing them into a DataFrame
### To loop through each playlist that are then returned by the API
### To convert the list of dictionaries into a DataFrame
### To check that the playlists table loaded correctly 

```python
playlists = sp.current_user_playlists()
playlist_data = []

for playlist in playlists['items']:
    playlist_data.append({
        "playlist_name": playlist['name'],
        "playlist_id": playlist['id']
    })
playlist_df = pd.DataFrame(playlist_data)

playlist_df
```

### This is to build a detailed DataFrame
### To loop over each playlist in playlist_df
### Requesting the tracks in the playlist
### A loop in place in case there are multiple pages of tracks
 ### To skip local files that do not have Spotify audio features
### To get the track ID and if it is missing, to skip it
### Try to request the audio of the playlist
 ### If spotify returns nothing then skip or return None.
 ### If the permission or rate limit is exceeded, then print the output and continue
### Save the track information and the selected audio features
### if there is another page of tracks, request for them, or otherwise, end the loop 
### Creating a dataframe with audio features

```python

from spotipy.exceptions import SpotifyException

tracks_list = []


for row in playlist_df.itertuples():
    playlist_id = row.playlist_id
    playlist_name = row.playlist_name

    results = sp.playlist_tracks(playlist_id)


    while results:
        for item in results['items']:
            track = item['track']
            if track is None:
                continue

   
            if track.get("is_local"):
                continue

            track_id = track.get('id')
            if track_id is None:
                continue

            try:
                af_list = sp.audio_features([track_id])
               
                if not af_list or af_list[0] is None:
                    
                    continue
                audio_features = af_list[0]
            except SpotifyException as e:
               
                print(f"Skipping track {track_id} ({track['name']}): {e}")
                continue
            tracks_list.append({
                "playlist_name": playlist_name,
                "track_name": track['name'],
                "artist_name": track['artists'][0]['name'],
                "track_id": track_id,
                "popularity": track['popularity'],
                "danceability": audio_features['danceability'],
                "energy": audio_features['energy'],
                "valence": audio_features['valence'],
                "tempo": audio_features['tempo'],
                "acousticness": audio_features['acousticness'],
                "instrumentalness": audio_features['instrumentalness'],
                "loudness": audio_features['loudness'],
            })

      
        if results['next']:
            results = sp.next(results)
        else:
            results = None

df = pd.DataFrame(tracks_list)
```

### Building a cleaner track-level Dataframe with the basic meta data for every song
### Looping through each playlist 
### Requests the track for a playlist
### A loop incase the playlist is too long
### Get the track ID and skip any weird/local tracks without an ID
### Storing the basic info about each track into the DataFrame
### To go to the next page of results 
### Creating the main DataFrame I use for analysis
### Checking the shape to see how many rows and columns I have

```python
from spotipy.exceptions import SpotifyException

track_rows = []

for row in playlist_df.itertuples():
    playlist_id = row.playlist_id
    playlist_name = row.playlist_name

    results = sp.playlist_tracks(playlist_id)
    while results:
        for item in results["items"]:
            track = item["track"]
            if track is None:
                continue
            track_id = track.get("id")
            if track_id is None:
                continue

            track_rows.append({
                "playlist_name": playlist_name,
                "track_name": track["name"],
                "artist_name": track["artists"][0]["name"],
                "track_id": track_id,
                "popularity": track.get("popularity"),
                "duration_ms": track.get("duration_ms"),
                "explicit": track.get("explicit"),
            })

        if results["next"]:
            results = sp.next(results)
        else:
            results = None

tracks_df = pd.DataFrame(track_rows)
tracks_df.shape
```
